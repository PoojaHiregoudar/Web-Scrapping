```python
import pandas as pd
import time
from multiprocessing import Pool
from urllib.request import urlopen
import requests, json, os, warnings, re
from bs4 import BeautifulSoup
import json


url = urlopen("https://www.cermati.com/karir")
url = requests.get("https://www.cermati.com/karir")

soup = BeautifulSoup(url.content, 'html.parser')

script = soup.find('script', type="application/json").text

pid = json.loads(script)
pi = pid['smartRecruiterResult']['all']['content']

l =[]
def gen_url():
    for i in range(len(pi)):
        l.append(pi[i]['ref'])
        
        
        

def scrap(l):
    df = pd.DataFrame(columns = ['Department', 'Title', 'Location','Description','Qualification'])
    for i in range(len(l)):
        domain = l[i]
        response = requests.get(domain)
        df = df.append({'Title':response.json()['name'],'Location':response.json()['location']['city'],'Description': response.json()['jobAd']['sections']['jobDescription']['text'],'Qualification':response.json()['jobAd']['sections']['qualifications']['text'],'Department': response.json()['department']['label']}, ignore_index = True)
    return df
```

```python
gen_url()
```

```python
p = Pool(10)
df = p.map(scrap(l), l)
p.terminate()
p.join()
```

```python
y = df.to_dict(orient ='records')
```

```python
from itertools import groupby
from operator import itemgetter

# dictionary
y = df.to_dict(orient ='records')


y = sorted(y,key = itemgetter('Department'))


for key, value in groupby(y,key = itemgetter('Department')):
    print(key)
    for k in value:
        print(k)
```

```python
## saving as a json

# Serializing json
json_object = json.dumps(y, indent=4)
 
# Writing to sample.json
with open("solution.json", "w") as outfile:
    outfile.write(json_object)
```
